```python
# Очистим текст от стоп слов и пунтуации 
def clear_text(text):
    #без пунтуации
    content = re.sub(r'[^\w\s]', '', text)
    tokens = nltk.tokenize.word_tokenize(content)
    #без стоп слов
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if not token.lower() in stop_words]
    return tokens
```

```python
#Поиск имен будем делать через единичные заглавные слова. В топе будут имена, так как имя + фамилия встречается реже

#Поиск имени и фамилии будем делать по встречаемости 2 заглавных слов рядом
#Можно пренебречь, что Гарри Поттер и Поттери Гарри будут разными n граммами, так как к счастью в английском принято писать сначало имя, а потом фамилию

#Поиск профессоров будем делать через встречаемость professor/ Professor + заглавное слово

# Очистим текст от стоп слов и пунтуации 
def find_names(text):
    #без пунтуации
    content = re.sub(r'[^\w\s]', '', text)
    tokens = nltk.tokenize.word_tokenize(content)
    #без стоп слов
    stop_words = set(stopwords.words('english'))
    tokens_names = [token for token in re.findall('[A-Z]{1}[a-z]{3,}',content) if token.lower() not in stop_words] #ограничим минимально 3 знаками, 
                                                                                #так как Ron - 3 знака, а меньше дает много мусора
    tokens_surnames = [token for token in re.findall('[A-Z]{1}[a-z]{3,}.[A-Z]{1}[a-z]{3,}',content)]
    tokens_professors = [token for token in re.findall('Professor.{1}[A-Z]{1}[a-z]{3,}|professor.{1}[A-Z]{1}[a-z]{3,}]',content)]
    return tokens_names,tokens_surnames,tokens_professors

```

## Применим сверточные сети

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
```

```python
df_train['words'] = df_train[2].apply(word_tokenize)
df_test['words'] = df_test[2].apply(word_tokenize)
df_dev['words'] = df_dev[2].apply(word_tokenize)
```

Идея в том, что мы обучаемся вдоль каждой размерности вектора слов

```python
class CustomDataset(Dataset):
    def __init__(self, train=True):
        if train:
            self.df = df_train
        else:
            self.df = df_dev

    def __len__(self):
        return len(self.df)

    def __getitem__(self, index):
        x = np.zeros(256 * 100).reshape(256, 100)
        for i, j in enumerate(list(self.df.loc[index, 'words'])):
            if i < 100:
                x[:, i] = model_surevised.get_word_vector(j)
            else:
                break
        vect = torch.tensor(x.astype(np.float32))
        return {'sample': vect,
            'label': self.df.loc[index, 'true_lables']}
```

```python
train_dataset = CustomDataset()
test_dataset = CustomDataset(False)
```

```python
class Model(nn.Module):
    def __init__(self):
        super().__init__()

        self.inner = nn.Sequential(
            nn.Conv1d(256, 100, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv1d(100, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Flatten(),
            nn.Linear(16*100, 256),
            nn.ReLU(),
            nn.Linear(256, 85)
            )


    def forward(self, x):
        return self.net(x)

    def compute_all(self, batch):  # удобно сделать функцию, в которой вычисляется лосс по пришедшему батчу
        x = batch['sample']
        y = batch['label']
        logits = self.inner(x)

        loss = F.cross_entropy(logits, y)
        acc = (logits.argmax(axis=1) == y).float().mean().cpu().numpy()
        metrics = dict(acc=acc)

        return loss, metrics
```

```python
# класс из семенаров по дл
class Trainer:
    def __init__(self, model, optimizer, train_dataset, val_dataset, batch_size=128):
        self.model = model
        self.optimizer = optimizer
        self.train_dataset = train_dataset
        self.val_dataset = val_dataset

        self.batch_size = batch_size
        self.val_losses = []
        self.device = 'cpu'
        if torch.cuda.is_available():
            self.device = torch.cuda.current_device()
            self.model = self.model.to(self.device)
        

    def train(self, num_epochs):
        model = self.model
        optimizer = self.optimizer
        
        train_loader = DataLoader(self.train_dataset, shuffle=True, pin_memory=True, batch_size=self.batch_size)
        val_loader = DataLoader(self.val_dataset, shuffle=False, pin_memory=True, batch_size=self.batch_size)
        best_loss = float('inf')
        
        for epoch in range(num_epochs):
            model.train()
            for batch in tqdm(train_loader):
                batch = {k: v.to(self.device) for k, v in batch.items()}
                loss, details = model.compute_all(batch)
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
            
            model.eval()
            
            for batch in tqdm(val_loader):
                batch = {k: v.to(self.device) for k, v in batch.items()}
                loss, details = model.compute_all(batch)
                self.val_losses.append(loss.item())
```

```python
import torch.optim as optim
```

```python
model = Model()
opt = optim.SGD(model.parameters(), lr=1e-3)
trainer = Trainer(model, opt, train_dataset, test_dataset, batch_size=128)
```

```python
trainer.train(10)
```

```python
def get_predictions(data):
    data = list(data)
    x = np.zeros(256 * 100).reshape(256, 100)
    for i, j in enumerate(data):
        if i < 100:
            x[:, i] = model_surevised.get_word_vector(j)
        else:
            break
        vect = torch.tensor(x.astype(np.float32)).unsqueeze(0)
        return model.inner(vect).argmax().item()
```
